### fillna with percentage < = 0.5 % with mean

```python
cols=final_df[['agricultural_land_sq_km',
                  'animals_production_head', 'cereal_yield_kg_per_hectare',
                  'co2_emissions_kt','methane_emissions_kt_of_co2_equivalent',
                  'nitrous_oxide_emissions_thousand_metric_tons_of_co2_equivalent','harvested_area_of_cereal_crops_ha',
                  'cereal_crops_production_ton','cereal_crops_yield_hg_ha','land_area_equipped_for_irrigation_ha']]

for col in final_df.columns:
    if col in cols:
        final_df[col]=final_df[col].fillna(final_df[col].mean())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot

%matplotlib inline
jtplot.style('gruvboxd')
```

```python
data = pd.read_csv('water_dataset_part_1.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize = (20, 10))
plt.title('Progression of water loss via climatic and evapotranspirative activities', fontsize = 30, pad = 20)
plt.plot(data['Year'], data['Climatic Water Deficit-Annual kg/m2'], label = 'Water Deficit via climate change', marker = 'D', markersize = 20)
plt.plot(data['Year'], data['WaPOR Reference Evapotranspiration-Annual mm'], label = 'Water lost via evapotranspiration', marker = 'D', markersize = 20)
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
plt.figure(figsize = (20, 10))
plt.title('Progression of water productivity', fontsize = 30, pad = 20)
plt.plot(data['Year'], data['WaPOR Gross Water Productivity-Annual kg/m3'], label = 'Gross water productivity', marker = 'D', markersize = 20)
plt.plot(data['Year'], data['WaPOR Net Primary Production-Annual g/m2'], label = 'Net water productivity', marker = 'D', markersize = 20)
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
plt.figure(figsize = (20, 10))
plt.title('Progression of water loss via climatic and evapotranspirative activities', fontsize = 30, pad = 20)
plt.plot(data['Year'], data['Climatic Water Deficit-Annual kg/m2'], label = 'Water loss via climatic changes', marker = 'D', markersize = 20)
plt.plot(data['Year'], data['Potential Evapotranspiration-Annual kg/m2'], label = 'Potential water loss via evapotranspirational activities', marker = 'D', markersize = 20)
plt.legend(loc = 0, fontsize = 'x-large')
plt.show()
```

From the dataset and the visualization above, it is evident that:
1. Water is lost at a far higher rate that it is produced.
2. The annual water productivity rates (both gross and net) have been in steady lapse, practically zero.
3. In the past year alone, water content has dropped significantly both from climate change and from evaporation-related activities. Evapotranspiration, howeever, is more of a threat as far as water loss is concerned.

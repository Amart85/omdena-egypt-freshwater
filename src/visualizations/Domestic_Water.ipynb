```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Excel/infonile-_-story-data-domestic-water-egypt.csv',sep=r'\s*,\s*',
                           header=0, encoding='ascii', engine='python')
df
```

```python
df.info()
```

### We can notice that fresh water from salted water *sea water* producing is so low, the total is 21.64 that is neglected beside surface and ground water

```python
SW = df["Surface Water - SW (MCM)"][27]
GW = df["Groundwater - GW (MCM)"][27]
SWD = df["Sea Water Desalination - SWD (MCM)"][27]
TOT = df['Total (MCM)'][27]
AREA = df['Area'][27]
```

```python
percentage_used=[]
for i in df["Total (MCM)"]:
    percentage_used.append(round(i/TOT,3))
    
Ground=[]
for i in df["Groundwater - GW (MCM)"]:
    Ground.append(round(i/GW,3))

Surface=[]
for i in df["Surface Water - SW (MCM)"]:
    Surface.append(round(i/SW,3))
    
Desalination=[]
for i in df["Sea Water Desalination - SWD (MCM)"]:
    Desalination.append(round(i/SWD,3))
    
Area_perc=[]
for i in df["Area"]:
    Area_perc.append(round(i/AREA,3))
```

```python
df['Total Percentage']=percentage_used
df['SW Percentage']=Surface
df['GW Percentage']=Ground
df['SWD Percentage']=Desalination
df['Area Percentage']=Area_perc

df
```

```python
df2 = df.drop(index=27,axis=0)
df2
```

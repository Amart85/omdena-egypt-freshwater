```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_excel("Irrigation usage in different areas in Egypt.xlsx",skiprows=[1])
```

```python
df
```

```python
df.drop("Ser No",axis=1,inplace=True)
```

```python
total=df["total (ha)"][26]
total_ground= df["with groundwater"][26]
total_surface= df["with surface water(ha)"][26]
```

```python
percentage_used=[]
for i in df["total (ha)"]:
    percentage_used.append(round(i/total,3))
    
ground=[]
for i in df["with groundwater"]:
    ground.append(round(i/total_ground,3))

surface=[]
for i in df["with surface water(ha)"]:
    surface.append(round(i/total_surface,3))
```

```python
df["ground_percentage"]=ground
df["surface_percentage"]=surface
df["total_percentage"]=percentage_used
```

```python
dfn=df.drop(index=26)
```

```python
dfn
```

```python
a4_dims = (12, 10)
fig, ax =plt.subplots(figsize=a4_dims)
ax = sb.barplot(data = dfn, x = 'ground_percentage', y = 'Governorate',)
plt.grid()
```

```python
a4_dims = (10, 10)
fig, ax =plt.subplots(figsize=a4_dims)
ax = sb.barplot(data = dfn, x = 'surface_percentage', y = 'Governorate',)
plt.grid()
```

```python
a4_dims = (10, 10)
fig, ax =plt.subplots(figsize=a4_dims)
ax = sb.barplot(data = dfn, x = 'total_percentage', y = 'Governorate',)

plt.grid()
```

### Now here is another prespective in the analysis with the considiration of the area in each governorate.

As it's conviniat that the more area is, the more water consumption. 

Therefore, we need to divide the water consumbtion over the area to see whether the governorate use the water in a right way or waste it. 

There are some governorate with 0 area. therefore, the percentage area fro them will be set to be 0.002. As the total some for the column is 0.99. Therefore 0.01 will be disteributed around the 5 governorate with 0 value. 

```python
for i in range(5): 
    dfn["Percent area"][i]=0.002
```

```python
total_area=[]
for i in range(len(dfn["total_percentage"])):
    total_area.append(round(dfn["total_percentage"][i]/dfn["Percent area"][i]))

ground_area=[]
for i in range(len(dfn["ground_percentage"])):
    ground_area.append(round(dfn["ground_percentage"][i]/dfn["Percent area"][i]))

surface_area=[]
for i in range(len(dfn["surface_percentage"])):
    surface_area.append(round(dfn["surface_percentage"][i]/dfn["Percent area"][i]))
```

```python
dfn["total_area"]=total_area
dfn["surface_area"]=surface_area
dfn["ground_area"]=ground_area
```

```python
a4_dims = (10, 10)
fig, ax =plt.subplots(figsize=a4_dims)
ax = sb.barplot(data = dfn, x = 'surface_area', y = 'Governorate',)

plt.grid()
```

**Here i t may  be not accurate as we do not know the exact percantage of area in the first 5 governate. However, we can see that Dumyate used huge amount of surface water according to its area.**

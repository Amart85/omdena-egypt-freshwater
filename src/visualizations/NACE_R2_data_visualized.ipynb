<h2>But what exactly is NACE_R2?</h2>

<a href="https://ec.europa.eu/eurostat/statistics-explained/index.php?title=Glossary:Statistical_classification_of_economic_activities_in_the_European_Community_(NACE)">The Statistical classification of economic activities in the European Community</a>, abbreviated as NACE, is the classification of economic activities in the European Union (EU); the term NACE is derived from the French Nomenclature statistique des activités économiques dans la Communauté européenne. Various NACE versions have been developed since 1970.

NACE R2 stands for the second revision of the classification of economic activities in the European Union. <br />

However, their purpose here is only to provide codes for different types of economic activitues.





We are only concerned with Egypt's data so we'll select the EG country code.

```python
df_eg = df[df['country'] == 'EG']
```

```python
df_eg
```

The MIO_M3 unit is <a href="https://unece.org/fileadmin/DAM/cefact/recommendations/add3lmdt.htm">millions of meters cubed</a>

The A, C, and EP_HH stand for "Agriculture, forestry and fishing", "Manufacturing", and "Households" respectively.<br />
The nomenclature can be obtained by searching <a href="https://ec.europa.eu/eurostat/ramon/nomenclatures/index.cfm?TargetUrl=LST_NOM_DTL_LINEAR&StrNom=CL_NACE2&StrLanguageCode=EN&IntCurrentPage=15">here</a>

```python
available_years = df_eg.columns[5:12]
```

```python
df_eg[available_years]
```

```python
df_eg = df_eg.copy()
for col in available_years:
    df_eg[col] = df_eg[col].astype(float)
```

While the data is small enough that it doesn't require visualization, let's see how it looks anyway.

```python
df_eg[df_eg.nace_r2 == 'A'][available_years[::-1]].transpose().plot(legend=False)
```

```python
df_eg[df_eg.nace_r2 == 'C'][available_years[::-1]].transpose().plot(legend=False)
```

```python
df_eg[df_eg.nace_r2 == 'EP_HH'][available_years[::-1]].transpose().plot(legend=False)
```

Note that while the numbers on the y-axis are similar in magnitude, the data itself is measured in millions, so an 0.1 difference can still be significant.

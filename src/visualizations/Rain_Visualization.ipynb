```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1. Import Data using Pandas

```python
data = pd.read_csv("../data/Final Datasets/Climate related datasets/weather_df.csv")
data.head()
```

# 2. Exploratory Data Analysis

```python
data.info()
```

## Rainfall on Egypt's Analysis

```python
print("The most Rainfall volume is: ", data['Rainfall(MM)'].mode()[0], "(MM)")
```

```python
sns.displot(data['Rainfall(MM)'] , kde=True, height=8.27, aspect=11.7/8.27)
```

```python
plt.figure(figsize = (10,5))
fig = sns.lineplot(x = 'Year', y = 'Rainfall(MM)',data = data)
```

```python
plt.figure(figsize = (10,5))
fig = sns.lineplot(x = 'Year', y = 'Rainfall(MM)',data = data[data['Year'] > 2015])
```

```python
# Average stock prize in last 5 years
tesla_stock = hand_washing.loc[hand_washing.units=='urban (% of urban population)'].value
ford_stock = hand_washing.loc[hand_washing.units=='rural (% of rural population)'].value

year = [2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010, 2009, 2008, 2007,
       2006, 2005, 2004, 2003, 2002, 2001, 2000]

# We cannot add width to year so we create another list
indices = np.arange(len(year))
fig = plt.figure(figsize = (10, 5))

width = 0.20

# Plotting
plt.bar(indices, tesla_stock, width=width)

# Offsetting by width to shift the bars to the right
plt.bar(indices + width, ford_stock, width=width)


# Displaying year on top of indices
plt.xticks(ticks=indices, labels=year)


plt.xlabel("year")
plt.ylabel("value")
plt.title("urban vs. rural")

plt.savefig("plot.png")
plt.show()
```

```python
indicator_fresh_water.loc[indicator_fresh_water.units=='agriculture (% of total freshwater withdrawal)'].drop_duplicates(subset=['year'])
```

```python
# Average stock prize in last 5 years
agri_stock = indicator_fresh_water.loc[indicator_fresh_water.units=='agriculture (% of total freshwater withdrawal)'].drop_duplicates(subset=['year']).value
domestic_stock = indicator_fresh_water.loc[indicator_fresh_water.units=='domestic (% of total freshwater withdrawal)'].value
industry_stock = indicator_fresh_water.loc[indicator_fresh_water.units=='industry (% of total freshwater withdrawal)'].value


year = [2017, 2012, 2007, 2002, 1997]

# We cannot add width to year so we create another list
indices = np.arange(len(year))
fig = plt.figure(figsize = (10, 5))

width = 0.20

# Plotting
plt.bar(indices, agri_stock, width=width)

# Offsetting by width to shift the bars to the right
plt.bar(indices + width, domestic_stock, width=width)

plt.bar(indices + 2*width, industry_stock, width=width)


# Displaying year on top of indices
plt.xticks(ticks=indices, labels=year)


plt.xlabel("year")
plt.ylabel("value")
plt.title("Domestic vs. Agriculture vs industry")

plt.savefig("plot.png")
plt.show()
```

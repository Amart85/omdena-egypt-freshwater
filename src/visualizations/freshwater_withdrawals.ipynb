```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv("freshwater_withdrawals.csv",index_col=[0])
```

```python
df
```

```python
df=df.drop(["country_iso3","country_name","indicator_code"],axis=1)
```

```python
indicators_name=list(df.indicator_name.unique())
```

```python
def data(indicators_name,da):
    indicators_name_data=da[da["indicator_name"]==indicators_name]
    return indicators_name_data
```

```python
d=[]
for i in indicators_name:
    d.append(data(i,df))
```

```python
dd=[]
for i in d:
    if i.units.nunique()==1:
        dd.append(i)
```

```python
for i in dd:
    a4_dims = (10, 10)
    fig, ax =plt.subplots(figsize=a4_dims)
    sb.barplot(data = i, x = 'year', y = 'value',hue="indicator_name")
    plt.title(list(i.units.unique())[0])

    plt.grid()
```
